```python
import pandas as pd
import numpy as np
df = pd.read_csv('C:\\Users\\tirumaleswarasai.t\\Desktop\\weather_data_1.csv')
df.set_index('day',inplace=True)
df
```

```python
new_df = df.replace(-99999,np.NaN) # it replaces -99999 with NaN present anywhere in the table
new_df
```

```python
# if we want to replace two differennt values then we will be using list
new_df = df.replace([-99999,-88888],np.NaN)
new_df
```

```python
# if we want to replace 0 in event colums we simply cant add 0 in the list above 
#as it will replace 0 not only in events column but in the whole table
# so we will be using dictionary for this purpose
new_df = df.replace({
    'temperature':-99999,
    'windspeed': [-88888,-99999],
    'event': '0'
},np.NaN)
new_df
```

```python
# but if we want to replace 0 in event column with any other value other than NaN
df_new = df.replace({
    -99999:np.NaN,
    -88888:np.NaN,
    '0': 'No Rain'
})
df_new
```

```python
# to remove any unwanted parameters or characters like below table
# we will be using regex
df_new = df.replace('[A-Za-z]','',regex=True)
df_new
```

```python
#but if we see in the above table all the characters in the event column also got replaced with whitespace
# so it is important to mention the column which we want to replace
new_df = df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
newd = new_df.replace({
    -99999: np.NaN,
    '0':'No fuckin rain'
})
newd
```
